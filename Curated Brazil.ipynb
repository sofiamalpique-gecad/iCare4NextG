We need to make a split so later we can validate our results.

````python
from sklearn.model_selection import train_test_split

# Let's say we want to split the data in 80:10:10 for train:valid:test dataset
train_size=0.8

X = df.drop(columns = ['Outcome']).copy()
y = df['Outcome']

# In the first step we will split the data in training and remaining dataset
X_train, X_rem, y_train, y_rem = train_test_split(X,y, train_size=0.8)

# Now since we want the valid and test size to be equal (10% each of overall data). 
# we have to define valid_size=0.5 (that is 50% of remaining data)
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=0.5)

print(X_train.shape), print(y_train.shape)
print(X_valid.shape), print(y_valid.shape)
print(X_test.shape), print(y_test.shape)
````

As we did want to reinforce the minority class, but at the same time, we didn't want to have all the entries repeated, so we are only making a copy of half of the minority class and then make a slice on the **majority** class of the size of ```1.5*(size_of_minority_class)```

# usar os 132k restantes para validação (então 8k representam corretamente as características do DS)

````python
df_survived = df_train.loc[df_train['Outcome'] == 0]
df_died = df_train.loc[df_train['Outcome'] == 1]

# Over-sampling:
died_subset = df_died.sample(frac=0.5)
died_subset = pd.concat([died_subset, df_died], axis=0)
size = len(died_subset)

#Under-sampling:
survived_subset = df_survived.sample(n=size)

even_df = pd.concat([survived_subset, died_subset])

# we should shuffle them in order to not have all 1s after 0s
even_df = even_df.sample(frac = 1, random_state = 1).reset_index()

del even_df['index']
````

````python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

# importing models
from sklearn import svm, tree
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import xgboost
import lightgbm as lgb
from sklearn.svm import LinearSVC
from sklearn.svm import NuSVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
````

````python
nb_classifier = GaussianNB()
svc_rbf_classifier = svm.SVC(kernel='rbf')
svc_linear_classifier = svm.SVC(kernel='linear')
svc_poly_classifier = svm.SVC(kernel='poly')
svc_sigmoid_classifier = svm.SVC(kernel='sigmoid')
svc_precomputed_classifier = svm.SVC(kernel='precomputed')
svm_classifier = make_pipeline(StandardScaler(),LinearSVC(random_state=0, tol=1e-5))
nusvc_classifier = make_pipeline(StandardScaler(), NuSVC())
rf_classifier = RandomForestClassifier(n_estimators=100)
adb_classifier = AdaBoostClassifier(n_estimators=100)
voting_classifier = VotingClassifier(estimators=[('adb', adb_classifier), ('rf', rf_classifier), ('nb', nb_classifier)], voting='hard')


classifiers = [svc_rbf_classifier, svc_linear_classifier, svc_poly_classifier, 
               svc_sigmoid_classifier, svc_precomputed_classifier, 
               svm_classifier, rf_classifier, adb_classifier, voting_classifier]
````

````python
X_train = even_df.loc[:, even_df.columns != 'Outcome']

y_train = even_df['Outcome']


for clf in classifiers:
  clf.fit(X_train,y_train)
  preds = clf.predict(X_valid)

  a1 = metrics.accuracy_score(y_valid,preds)
  a2 = metrics.f1_score(y_valid,preds)
  a3 = metrics.precision_score(y_valid,preds)
  a4 = metrics.recall_score(y_valid,preds)

  cm = metrics.confusion_matrix(y_valid, preds)

  

  print()
  print('Scores with Validation set ONLY:')
  print('{} accuracy:'.format(clf),a1)
  print('{} precision:'.format(clf),a2)
  print('{} recall:'.format(clf),a3)
  print('{} f1:'.format(clf),a4)
  print('{} confusion matrix:'.format(clf))
  print(cm)

  print()
  print()
````
