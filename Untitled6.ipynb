```python
import matplotlib.pyplot as plt

# Define the data
clusters = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8']
percentages = [46.95, 6.62, 18.17, 0.78, 6.98, 7.2, 0.95, 8.48, 3.87]

# Define colors for each cluster
cluster_colors = {
    'C0': '#2FA173',
    'C2': '#2FA173',
    'C7': '#2FA173',
    'C3': '#BEFFE5',
    'C6': '#BEFFE5',
    'C8': '#BEFFE5',
}

# Plot circles with corresponding labels and percentages
fig, ax = plt.subplots(figsize=(10, 10))
for i, (cluster, percent) in enumerate(zip(clusters, percentages)):
    color = cluster_colors.get(cluster, '#6ED5AB')
    circle = plt.Circle((i, 0.5), (percent / 100), facecolor=color, edgecolor=None, clip_on=False)
    ax.add_artist(circle)
    if cluster == 'C0':
        plt.annotate(f'{cluster}\n{percent}%', xy=(i, 0.5), ha='center', va='center', color='black', fontsize=9)
    else:
        plt.annotate(f'{percent}%', xy=(i, 0.5 - (percent / 100) - 0.08), ha='center', va='center', color='black', fontsize=9)

# Set the aspect ratio to be equal, so the circles look like circles
ax.set_aspect('equal', adjustable='box')

# Set x and y axis labels
plt.xlabel('Clusters')
plt.ylabel('Percentage \n of Patients', labelpad=50)  # Adjust labelpad to add space between label and clusters

# Set title and remove y ticks
plt.title('Distribution of COVID-19 Comorbidities Clusters', pad=20)  # Adjust pad to add space between title and clusters
plt.xticks(range(len(clusters)), clusters, rotation=45)
plt.yticks([])

# Remove axis spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove ticks
ax.tick_params(bottom=False, left=False)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler

# Create a dataset with a 3:1 imbalance
X, y = make_classification(n_classes=2, class_sep=2,
weights=[0.3, 0.7], n_informative=3, n_redundant=1, flip_y=0,
n_features=20, n_clusters_per_class=1, n_samples=1000, random_state=10)

# Visualize the dataset
plt.hist(y, bins=np.arange(3)-0.5, rwidth=0.5, color='blue', alpha=0.7)
plt.title('Original Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1])
plt.show()

# Apply the random under-sampling
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

# Visualize the new dataset
plt.hist(y_res, bins=np.arange(3)-0.5, rwidth=0.5, color='green', alpha=0.7)
plt.title('Distribution after Random Under Sampling')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0, 1])
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.under_sampling import RandomUnderSampler

# Create a dataset with a 96:4 imbalance
X, y = make_classification(n_classes=2, class_sep=2,
weights=[0.96, 0.04], n_informative=3, n_redundant=1, flip_y=0,
n_features=20, n_clusters_per_class=1, n_samples=1000, random_state=10)

# Visualize the dataset
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Original Distribution')

# Apply the random under-sampling
rus = RandomUnderSampler()
X_res, y_res = rus.fit_resample(X, y)

# Visualize the new dataset
plt.subplot(1, 2, 2)
plt.scatter(X_res[:, 0], X_res[:, 1], c=y_res, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Distribution after Random Under Sampling')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE

# Create a dataset with a 96:4 imbalance
X, y = make_classification(n_classes=2, class_sep=2,
weights=[0.96, 0.04], n_informative=3, n_redundant=1, flip_y=0,
n_features=20, n_clusters_per_class=1, n_samples=1000, random_state=10)

# Visualize the dataset
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Original Distribution')

# Apply the SMOTE over-sampling
smote = SMOTE()
X_res, y_res = smote.fit_resample(X, y)

# Visualize the new dataset
plt.subplot(1, 2, 2)
plt.scatter(X_res[:, 0], X_res[:, 1], c=y_res, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Distribution after SMOTE')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.combine import SMOTEENN

# Create a dataset with a 96:4 imbalance
X, y = make_classification(n_classes=2, class_sep=2,
weights=[0.96, 0.04], n_informative=3, n_redundant=1, flip_y=0,
n_features=20, n_clusters_per_class=1, n_samples=1000, random_state=10)

# Visualize the dataset
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Original Distribution')

# Apply the SMOTEENN
smote_enn = SMOTEENN(random_state=0)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)

# Visualize the new dataset
plt.subplot(1, 2, 2)
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, alpha=0.5, edgecolor='k', cmap='Blues')
plt.title('Distribution after SMOTEENN')

plt.tight_layout()
plt.show()
```
